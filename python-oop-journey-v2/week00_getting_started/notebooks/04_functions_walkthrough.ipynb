# 🔧 Functions Walkthrough

## Writing Reusable Code

Functions are the building blocks of programming. They let you write code once and use it many times!

> "The only way to write complex software that works is to make it out of simple parts that work."

## 🤔 What is a Function?

A function is like a recipe:
1. You give it **inputs** (ingredients)
2. It **processes** them (cooking)
3. It gives you **output** (the dish)

```
┌─────────────────────────────┐
│  Input → Function → Output  │
│                             │
│  3, 5  →  add()   →   8     │
│  "Bob" → greet() → "Hi Bob!"│
└─────────────────────────────┘
```

## ✍️ Defining Functions

Use the `def` keyword to create a function:

````python
# Simple function with no inputs
def say_hello():
    print("Hello! 👋")

# Call the function
say_hello()
say_hello()
say_hello()  # Reuse as many times as you want!
````

---

## 📥 Parameters and Arguments

- **Parameters** are what the function expects
- **Arguments** are what you pass in

````python
# Function with one parameter
def greet(name):  # 'name' is a parameter
    print(f"Hello, {name}!")

greet("Alice")   # "Alice" is an argument
greet("Bob")     # "Bob" is an argument
greet("World")   # "World" is an argument

# Function with multiple parameters
def introduce(name, age, city):
    print(f"Hi, I'm {name}, {age} years old, from {city}.")

introduce("Carol", 25, "NYC")
introduce("David", 30, "LA")
````

---

## 📤 Return Values

Functions can send data back using `return`:


````python
# Function that returns a value
def add(a, b):
    result = a + b
    return result

# Capture the return value
sum_result = add(5, 3)
print(f"5 + 3 = {sum_result}")

# Use directly in expressions
print(f"10 + 20 = {add(10, 20)}")

# Multiple return values
def divide_with_remainder(dividend, divisor):
    quotient = dividend // divisor
    remainder = dividend % divisor
    return quotient, remainder  # Returns a tuple!

q, r = divide_with_remainder(17, 5)
print(f"\n17 ÷ 5 = {q} remainder {r}")
````

---

## 🎯 Default Parameters

Give parameters default values so they're optional:

````python
def greet(name, greeting="Hello", punctuation="!"):
    print(f"{greeting}, {name}{punctuation}")

# Use defaults
greet("Alice")

# Override some defaults
greet("Bob", "Hi")
greet("Carol", punctuation="?")

# Override all
greet("David", "Hey there", "...")

# Keyword arguments (order doesn't matter!)
greet(punctuation="!!!", name="Eve", greeting="Yo")
````

## 🌐 Variable Scope

Scope determines where variables can be accessed:

```
┌─────────────────────────────────────┐
│  Global Scope (accessible everywhere)│
│  message = "Hello"                  │
│                                     │
│  ┌─────────────────────────────┐   │
│  │ Function Scope (local)      │   │
│  │ x = 5  # Only inside here   │   │
│  └─────────────────────────────┘   │
└─────────────────────────────────────┘
```

````python
# Global variable
global_message = "I'm global!"

def demo_scope():
    local_message = "I'm local!"
    print(f"Inside function: {local_message}")
    print(f"Inside function: {global_message}")  # Can access global

demo_scope()
print(f"Outside function: {global_message}")
# print(local_message)  # ERROR! Can't access local variable outside function

# Shadowing (local takes precedence)
name = "Global"
def shadow_demo():
    name = "Local"  # This is a different variable!
    print(f"Inside: {name}")

shadow_demo()
print(f"Outside: {name}")  # Still "Global"
````

## 📦 *args and **kwargs

Handle variable numbers of arguments:

````python
# *args - variable positional arguments
def sum_all(*numbers):
    total = 0
    for n in numbers:
        total += n
    return total

print(f"sum_all(1, 2, 3) = {sum_all(1, 2, 3)}")
print(f"sum_all(10, 20) = {sum_all(10, 20)}")
print(f"sum_all() = {sum_all()}")

# **kwargs - variable keyword arguments
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"  {key}: {value}")

print("\nPerson info:")
print_info(name="Alice", age=25, city="NYC")

print("\nProduct info:")
print_info(name="Laptop", price=999.99, brand="TechCo")
````

## 📖 Docstrings: Documenting Your Functions

Good functions explain what they do:

````python
def calculate_area(length, width):
    """
    Calculate the area of a rectangle.
    
    Args:
        length (float): The length of the rectangle
        width (float): The width of the rectangle
    
    Returns:
        float: The area of the rectangle
    
    Example:
        >>> calculate_area(5, 3)
        15
    """
    return length * width

# Access the documentation
help(calculate_area)

# Use the function
print(f"\nArea: {calculate_area(5, 3)}")
````

## 🔨 Lambda Functions: Quick One-Liners

For simple functions, use lambda (anonymous) syntax:

````python
# Regular function
def square(x):
    return x ** 2

# Equivalent lambda
square_lambda = lambda x: x ** 2

print(f"square(5) = {square(5)}")
print(f"square_lambda(5) = {square_lambda(5)}")

# Multiple arguments
add = lambda a, b: a + b
print(f"\nadd(3, 4) = {add(3, 4)}")

# Common use case: sorting
students = [
    ("Alice", 85),
    ("Bob", 92),
    ("Charlie", 78)
]

# Sort by score (second element)
sorted_students = sorted(students, key=lambda x: x[1], reverse=True)
print(f"\nSorted by score: {sorted_students}")
````
