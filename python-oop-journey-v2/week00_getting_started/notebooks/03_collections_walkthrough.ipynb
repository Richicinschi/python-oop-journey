# 📦 Collections Walkthrough

## Lists, Dictionaries, Sets, and Tuples

Collections are Python's way of storing multiple items together. Think of them as containers that can hold any number of things!

---

## 📋 Lists: Ordered Collections

Lists are like ordered sequences - think of a shopping list or a playlist.

```
Visual representation:
┌─────┬─────┬─────┬─────┬─────┐
│  0  │  1  │  2  │  3  │  4  │  ← Index (position)
├─────┼─────┼─────┼─────┼─────┤
│ apple│banan│cherr│date │elder│  ← Value
└─────┴─────┴─────┴─────┴─────┘
```

````python
# Creating lists
fruits = ["apple", "banana", "cherry"]
numbers = [1, 2, 3, 4, 5]
mixed = ["hello", 42, 3.14, True]

print("Fruits:", fruits)
print("Numbers:", numbers)
print("Mixed types:", mixed)

# Empty list
empty = []
also_empty = list()

# List length
print(f"\nNumber of fruits: {len(fruits)}")
````

---

### Accessing List Items

````python
colors = ["red", "green", "blue", "yellow", "purple"]

# Access by index (starts at 0!)
print(f"First color: {colors[0]}")
print(f"Second color: {colors[1]}")
print(f"Third color: {colors[2]}")

# Negative indices (count from end)
print(f"\nLast color: {colors[-1]}")
print(f"Second to last: {colors[-2]}")

# Slicing [start:end]
print(f"\nFirst three: {colors[0:3]}")
print(f"Last two: {colors[-2:]}")
print(f"Middle ones: {colors[1:4]}")
print(f"Every other: {colors[::2]}")  # Step by 2
````

---

### Modifying Lists

````python
# Lists are mutable (can be changed)
todo = ["wake up", "eat breakfast"]
print(f"Initial: {todo}")

# Add items
todo.append("go to work")  # Add to end
print(f"After append: {todo}")

todo.insert(1, "brush teeth")  # Insert at position
print(f"After insert: {todo}")

todo.extend(["lunch", "meeting"])  # Add multiple
print(f"After extend: {todo}")

# Remove items
todo.remove("go to work")  # Remove by value
print(f"After remove: {todo}")

done = todo.pop(0)  # Remove and return at index
print(f"Popped '{done}', remaining: {todo}")

# Change items
todo[0] = "shower"
print(f"After change: {todo}")
````

---

### List Methods and Operations

````python
numbers = [3, 1, 4, 1, 5, 9, 2, 6]

# Sorting
numbers.sort()
print(f"Sorted: {numbers}")

numbers.reverse()
print(f"Reversed: {numbers}")

# Finding
print(f"\nCount of 1: {numbers.count(1)}")
print(f"Index of 5: {numbers.index(5)}")

# Check existence
print(f"\nIs 7 in list? {7 in numbers}")
print(f"Is 5 in list? {5 in numbers}")

# Copy (important!)
original = [1, 2, 3]
copy = original.copy()  # Or list(original) or original[:]
copy.append(4)
print(f"\nOriginal: {original}")
print(f"Copy: {copy}")
````

## 🗂️ Dictionaries: Key-Value Pairs

Dictionaries store data as key-value pairs, like a real dictionary (word → definition).

```
Visual representation:
┌─────────────┬─────────────┐
│    Key      │    Value    │
├─────────────┼─────────────┤
│  "name"    │  "Alice"   │
│  "age"     │     25      │
│  "city"    │  "NYC"     │
└─────────────┴─────────────┘
```

````python
# Creating dictionaries
person = {
    "name": "Alice",
    "age": 25,
    "city": "New York"
}

print("Person:", person)

# Access values
print(f"Name: {person['name']}")
print(f"Age: {person['age']}")

# Using get() (safer - won't error if key missing)
print(f"\nCountry: {person.get('country', 'Unknown')}")

# Check if key exists
print(f"\nHas 'name'? {'name' in person}")
print(f"Has 'job'? {'job' in person}")
````

---

### Modifying Dictionaries

````python
# Add or update
person["job"] = "Engineer"  # Add new key
person["age"] = 26  # Update existing
print(f"After updates: {person}")

# Remove
del person["city"]
print(f"After del: {person}")

job = person.pop("job")  # Remove and return
print(f"Popped job: {job}, remaining: {person}")

# Get all keys, values, or items
person = {"name": "Bob", "age": 30, "city": "LA"}
print(f"\nKeys: {list(person.keys())}")
print(f"Values: {list(person.values())}")
print(f"Items: {list(person.items())}")
````

## 🔗 Tuples: Immutable Sequences

Tuples are like lists, but **cannot be changed** after creation. They're great for fixed data.

```
Coordinates: (x, y)
RGB Color: (255, 128, 0)
Date: (2024, 3, 15)
```

````python
# Creating tuples
coordinates = (10, 20)
rgb = (255, 128, 0)
single_item = (42,)  # Note the comma!

print(f"Coordinates: {coordinates}")
print(f"RGB: {rgb}")

# Access like lists
print(f"\nX: {coordinates[0]}")
print(f"Y: {coordinates[1]}")

# Unpacking - super useful!
x, y = coordinates
print(f"\nUnpacked: x={x}, y={y}")

red, green, blue = rgb
print(f"Colors: R={red}, G={green}, B={blue}")

# Tuples are immutable (can't change)
# coordinates[0] = 5  # This would cause an error!
````

## 🎯 Sets: Unique Collections

Sets store only unique items - no duplicates allowed!

```
Before: [1, 2, 2, 3, 3, 3]
After:  {1, 2, 3}
```

````python
# Creating sets
numbers = {1, 2, 3, 3, 3, 3}  # Duplicates removed!
print(f"Set: {numbers}")

# From a list
letters = set(["a", "b", "a", "c", "b"])
print(f"Letters: {letters}")

# Set operations
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(f"\nSet A: {a}")
print(f"Set B: {b}")
print(f"Union (A | B): {a | b}")  # All items
print(f"Intersection (A & B): {a & b}")  # Common items
print(f"Difference (A - B): {a - b}")  # In A but not B
print(f"Symmetric Diff (A ^ B): {a ^ b}")  # In either but not both
````

## 🔄 Iterating Over Collections

````python
# List iteration
fruits = ["apple", "banana", "cherry"]
print("Fruits:")
for fruit in fruits:
    print(f"  🍎 {fruit}")

# Dictionary iteration
scores = {"Alice": 95, "Bob": 87, "Charlie": 92}
print("\nScores:")
for name, score in scores.items():
    print(f"  {name}: {score}")

# With index
print("\nNumbered fruits:")
for i, fruit in enumerate(fruits, 1):
    print(f"  {i}. {fruit}")

# List comprehension (powerful Python feature!)
squares = [x**2 for x in range(1, 6)]
print(f"\nSquares: {squares}")

# Dictionary comprehension
doubled = {k: v*2 for k, v in scores.items()}
print(f"Doubled scores: {doubled}")
````

## 📚 Quick Reference

| Collection | Syntax | Mutable | Ordered | Duplicates |
|------------|--------|---------|---------|------------|
| List | `[1, 2, 3]` | ✅ Yes | ✅ Yes | ✅ Yes |
| Tuple | `(1, 2, 3)` | ❌ No | ✅ Yes | ✅ Yes |
| Dict | `{"a": 1}` | ✅ Yes | ❌ No* | ❌ No (keys) |
| Set | `{1, 2, 3}` | ✅ Yes | ❌ No | ❌ No |

*As of Python 3.7+, dicts preserve insertion order

```python
# Common operations
len(collection)     # Get size
x in collection     # Check membership
for x in collection # Iterate

# List methods
list.append(x)      # Add to end
list.insert(i, x)   # Insert at index
list.remove(x)      # Remove first x
list.pop(i)         # Remove at index
list.sort()         # Sort in place
list.copy()         # Make a copy

# Dict methods
dict.keys()         # Get keys
dict.values()       # Get values
dict.items()        # Get pairs
dict.get(k, default) # Safe access
```
