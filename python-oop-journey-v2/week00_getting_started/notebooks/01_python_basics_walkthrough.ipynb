## 📦 Variables: Your Data Containers

Think of variables as labeled boxes where you store information:

```
┌─────────────────┐
│  name = "Bob" │  ← A box labeled 'name' containing "Bob"
└─────────────────┘

┌─────────────────┐
│  age = 25       │  ← A box labeled 'age' containing 25
└─────────────────┘
```

````python
# Creating variables is easy!
name = "Alice"
age = 30
height = 5.6
is_student = True

print("Name:", name)
print("Age:", age)
print("Height:", height, "feet")
print("Is student?", is_student)
````

---

## 🔍 Data Types: What Kind of Data?

Python automatically figures out what type of data you're storing:

| Type | Description | Examples |
|------|-------------|----------|
| `str` | Text (string) | `"Hello"`, `'World'` |
| `int` | Whole numbers | `42`, `-7`, `0` |
| `float` | Decimal numbers | `3.14`, `-0.5` |
| `bool` | True/False | `True`, `False` |

Let's check the types:

````python
# Use type() to check what type something is
print("Type of 'Hello':", type("Hello"))
print("Type of 42:", type(42))
print("Type of 3.14:", type(3.14))
print("Type of True:", type(True))

# You can also check variables
score = 100
print("\nType of score:", type(score))
````

---

## 🔤 Working with Strings (Text)

Strings are how Python handles text. You can use single quotes `' '` or double quotes `" "`.

````python
# Creating strings
greeting = "Hello"
name = 'Python'

# String concatenation (joining)
message = greeting + ", " + name + "!"
print(message)

# String repetition
print("Ha" * 3)  # Prints: HaHaHa

# String length
print("Length of 'Python':", len("Python"))

# f-strings (formatted strings) - modern and convenient!
age = 25
print(f"I am {age} years old")
print(f"Next year I'll be {age + 1}")
````

## 🧮 Operators: Making Things Happen

### Arithmetic Operators

```
+   Addition       5 + 3 → 8
-   Subtraction    5 - 3 → 2
*   Multiplication 5 * 3 → 15
/   Division       5 / 3 → 1.666...
//  Floor Division  5 // 3 → 1
%   Modulo         5 % 3 → 2 (remainder)
**  Power          5 ** 3 → 125
```

````python
# Arithmetic examples
a = 17
b = 5

print(f"{a} + {b} = {a + b}")
print(f"{a} - {b} = {a - b}")
print(f"{a} * {b} = {a * b}")
print(f"{a} / {b} = {a / b}")
print(f"{a} // {b} = {a // b} (whole number division)")
print(f"{a} % {b} = {a % b} (remainder)")
print(f"{a} ** {b} = {a ** b} (power)")

# Practical example: Pizza slices
pizzas = 4
slices_per_pizza = 8
total_slices = pizzas * slices_per_pizza
print(f"\n{pizzas} pizzas × {slices_per_pizza} slices = {total_slices} slices total! 🍕")
````

---

### Comparison Operators

These compare values and return `True` or `False`:

```
==  Equal to         5 == 5 → True
!=  Not equal to     5 != 3 → True
<   Less than        5 < 3 → False
>   Greater than     5 > 3 → True
<=  Less or equal    5 <= 5 → True
>=  Greater or equal 5 >= 3 → True
```

````python
# Comparison examples
x = 10
y = 20

print(f"x = {x}, y = {y}")
print(f"x == y: {x == y}")
print(f"x != y: {x != y}")
print(f"x < y: {x < y}")
print(f"x > y: {x > y}")
print(f"x <= 10: {x <= 10}")

# Compare strings too!
print(f"\n'apple' == 'apple': {'apple' == 'apple'}")
print(f"'apple' < 'banana': {'apple' < 'banana'} (alphabetical order)")
````

---

### Assignment Operators

Shorthand ways to update variables:

````python
score = 100
print(f"Starting score: {score}")

score += 10  # Same as: score = score + 10
print(f"After += 10: {score}")

score -= 5   # Same as: score = score - 5
print(f"After -= 5: {score}")

score *= 2   # Same as: score = score * 2
print(f"After *= 2: {score}")

score //= 3  # Same as: score = score // 3
print(f"After //= 3: {score}")
````

## 🔄 Type Conversion

Sometimes you need to convert one type to another:

````python
# Converting types

# String to integer
age_str = "25"
age_int = int(age_str)
print(f"'{age_str}' as integer: {age_int}")
print(f"Next year: {age_int + 1}")

# Integer to string
num = 42
num_str = str(num)
print(f"\n{num} as string: '{num_str}'")
print(f"Can concatenate: '{num_str} is the answer'")

# Float and int conversion
pi = 3.14159
print(f"\npi = {pi}")
print(f"int(pi) = {int(pi)} (truncates decimal)")
print(f"float(5) = {float(5)}")
````
