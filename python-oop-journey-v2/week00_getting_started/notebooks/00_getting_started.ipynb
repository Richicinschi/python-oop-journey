## 🔍 Understanding Errors

Errors are **learning opportunities**! Don't worry when you see them - they're just Python's way of saying "I need a little clarification."

Let's look at a common error:

````python
# This will cause an error - but that's okay!
# Run it to see what happens, then fix it

print("Hello World)  # Oops! Missing closing quote
````

**See the error?** It says `SyntaxError` because we forgot the closing quote.

🔧 **How to fix it:** Add the missing quote at the end: `"Hello World"`

Errors usually tell you:
- **What** went wrong (e.g., `SyntaxError`)
- **Where** it happened (line number)

## 📚 Variables: Storing Information

Variables are like labeled boxes where you store information:

````python
# Creating variables
name = "Alice"
age = 25
favorite_number = 42

# Using variables
print("Name:", name)
print("Age:", age)
print("Favorite number:", favorite_number)

# You can do math with variables
next_year = age + 1
print("Next year I'll be:", next_year)
````
