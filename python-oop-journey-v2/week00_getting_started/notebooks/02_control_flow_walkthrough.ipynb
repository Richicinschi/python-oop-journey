# 🔄 Control Flow Walkthrough

## Making Decisions and Repeating Actions

Control flow is how your program makes decisions and repeats tasks. It's like the director of a movie, deciding what happens next!

---

## 🤔 If Statements: Making Decisions

If statements let your code make choices:

```
if condition:
    do something
```

Think of it like this:
- IF it's raining → take an umbrella ☔
- IF you're hungry → eat something 🍕
- IF password is correct → grant access 🔓

````python
# Basic if statement
age = 18

if age >= 18:
    print("You are an adult!")
    print("You can vote.")

print("This always prints")  # Not indented, so always runs
````

**Important:** Indentation matters in Python! Use 4 spaces (or 1 Tab) for each level.

---

## 🔀 If-Else: Two Paths

What if you want to do one thing OR another?

```
if condition:
    do this
else:
    do that
```

````python
temperature = 15

if temperature > 20:
    print("It's warm! Wear a t-shirt.")
else:
    print("It's chilly! Bring a jacket.")

# Another example
score = 75

if score >= 60:
    print("You passed! 🎉")
else:
    print("You need to study more. 📚")
````

---

## 🔀🔀 If-Elif-Else: Multiple Paths

What if you have more than two options?

```
if condition1:
    do A
elif condition2:
    do B
elif condition3:
    do C
else:
    do D
```

````python
# Grade calculator
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Score: {score}, Grade: {grade}")

# Time of day greeting
hour = 14  # 2 PM in 24-hour format

if hour < 12:
    greeting = "Good morning!"
elif hour < 17:
    greeting = "Good afternoon!"
elif hour < 21:
    greeting = "Good evening!"
else:
    greeting = "Good night!"

print(greeting)
````

## 🔄 Loops: Doing Things Repeatedly

### For Loops: Counted Repetition

Use a `for` loop when you know how many times to repeat:

````python
# Count from 1 to 5
print("Counting up:")
for i in range(1, 6):
    print(f"  {i}")

# range() creates a sequence of numbers
# range(start, stop) goes from start to stop-1

print("\nCounting by 2s:")
for i in range(0, 11, 2):  # start, stop, step
    print(f"  {i}")

print("\nCounting down:")
for i in range(5, 0, -1):
    print(f"  {i}...")
print("Blast off! 🚀")
````

---

### Looping Through Collections

````python
# Loop through a list
fruits = ["apple", "banana", "cherry"]

print("My favorite fruits:")
for fruit in fruits:
    print(f"  🍎 {fruit}")

# Loop through a string
print("\nLetters in 'Python':")
for letter in "Python":
    print(f"  {letter}")

# Loop with index using enumerate
print("\nNumbered list:")
for index, fruit in enumerate(fruits, 1):
    print(f"  {index}. {fruit}")
````

---

### While Loops: Conditional Repetition

Use a `while` loop when you don't know how many times to repeat:

```
while condition:
    keep doing this
```

````python
# Simple while loop
count = 0

print("While loop counting:")
while count < 5:
    print(f"  Count is: {count}")
    count += 1  # Important: update the condition!

print("\nDone!")

# Practical example: doubling
value = 1
print("\nDoubling until we pass 100:")
while value <= 100:
    print(f"  {value}")
    value *= 2
````

⚠️ **Warning:** Make sure your while loop will eventually end! Otherwise, you'll have an infinite loop.

## 🛑 Loop Control: Breaking and Continuing

- `break` - Exit the loop immediately
- `continue` - Skip to the next iteration

````python
# break example: find first even number
print("Finding first even number > 10:")
for num in range(1, 20):
    if num > 10 and num % 2 == 0:
        print(f"  Found it: {num}")
        break  # Stop looking!

# continue example: skip even numbers
print("\nOdd numbers from 1 to 10:")
for num in range(1, 11):
    if num % 2 == 0:
        continue  # Skip even numbers
    print(f"  {num}")
````

## 🧩 Nested Structures

You can put loops inside loops, and if statements inside loops!

````python
# Multiplication table
print("Multiplication table (1-5):")
print("  ", end="")
for i in range(1, 6):
    print(f"{i:4}", end="")
print("\n" + "-" * 25)

for i in range(1, 6):
    print(f"{i} |", end="")
    for j in range(1, 6):
        print(f"{i*j:4}", end="")
    print()  # New line after each row

# Finding primes
print("\nPrime numbers from 2 to 20:")
for num in range(2, 21):
    is_prime = True
    for divisor in range(2, int(num ** 0.5) + 1):
        if num % divisor == 0:
            is_prime = False
            break
    if is_prime:
        print(f"  {num}")
````
